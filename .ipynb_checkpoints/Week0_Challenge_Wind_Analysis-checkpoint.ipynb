```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def process_and_analyze(data, location_name, variables=None):
    """
    Processes and analyzes solar radiation data for a specific location.

    Args:
        data (pandas.DataFrame): The solar radiation data.
        location_name (str): The name of the location.
        variables (list, optional): A list of variable names to analyze.
            If None, all numeric columns are analyzed. Defaults to None.

    Returns:
        None
    """

    # Error handling for missing files
    try:
        data['Timestamp'] = pd.to_datetime(data['Timestamp'])
        data.set_index('Timestamp', inplace=True)
    except KeyError:
        print(f"Error processing data for {location_name}: 'Timestamp' column not found.")
        return

    # Calculate summary statistics for numeric columns (if any)
    if data.select_dtypes(include=[np.number]).empty:
        print(f"No numeric columns found in data for {location_name}.")
    else:
        summary_stats = data.describe(include=variables)
        print(f"\nSummary statistics for {location_name}:")
        print(summary_stats.to_string())  # Improved output formatting

    # Analyze time-series trends (default or user-specified variables)
    if not variables:
        variables = ['GHI', 'DNI', 'DHI', 'Tamb']
    fig, ax = plt.subplots(figsize=(12, 6))
    data[variables].plot(kind='line', ax=ax)  # Concise plotting using pandas methods
    plt.title(f"Time-series trends for {location_name}")
    plt.xlabel("Time")
    plt.ylabel("Value")
    plt.legend(variables)
    plt.tight_layout()  # Adjust spacing for better readability
    plt.show()

    # Conduct correlation analysis
    correlation_matrix = data.corr()
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.imshow(correlation_matrix, cmap='coolwarm')  # Matplotlib imshow for heatmap
    ax.set_xticks(np.arange(len(correlation_matrix.columns)))
    ax.set_yticks(np.arange(len(correlation_matrix.columns)))
    ax.set_xticklabels(correlation_matrix.columns, rotation=45, ha='left')
    ax.set_yticklabels(correlation_matrix.columns)
    for (i, j), z in np.ndenumerate(correlation_matrix):
        ax.text(j, i, "{:0.2f}".format(z), ha='center', va='center', fontsize=8)  # Customized annotations
    plt.title(f"Correlation Matrix for {location_name}")
    plt.colorbar()
    plt.tight_layout()
    plt.show()

    # Create visualizations to identify patterns or anomalies
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.scatter(data['Tamb'], data['GHI'], c=data['Precipitation'], cmap='viridis', s=20)
    plt.title(f"Scatter Plot of Tamb vs. GHI with Precipitation for {location_name}")
    plt.xlabel("Tamb (°C)")
    plt.ylabel("GHI (W/m²)")
    plt.colorbar(label='Precipitation')
    plt.tight_layout()
    plt.show()

    # Box plot to detect anomalies
    fig, ax = plt.subplots(figsize=(12, 6))
    data[variables].plot(kind='box', ax=ax)
    plt.title(f"Box Plot for {location_name}")
    plt.tight_layout()
    plt.show()

    # Histogram for distribution of GHI
    fig, ax = plt.subplots(figsize=(12, 6))
    data['GHI'].hist(bins=20, density=True, kde=True, ax=ax)  # More informative histogram
    plt.title(f"Histogram of GHI for {location_name}")
    plt.xlabel("GHI (W/m²)")
    plt.ylabel("Density")  # Adjusted label for density plots
    plt.show()


    # Clean the dataset by addressing anomalies or missing values
    # Replace negative values with NaN and drop them
    data.replace(-1, np.nan, inplace=True)
    data.dropna(inplace=True)

    # Check missing values after cleaning
    missing_values = data.isnull().sum()
    print(f"\nMissing values after cleaning for {location_name}:")
    print(missing_values)
```

```python
# Load data from CSV files (adjust the file paths)
try:
    benin_data = pd.read_csv("benin-malanville.csv")
    sierra_leone_data = pd.read_csv("sierraleone-bumbuna.csv")
    togo_data = pd.read_csv("togo-dapaong_qc.csv")
except FileNotFoundError:
    print("Error: One or more CSV files not found. Please check file paths.")
    exit()
```

```python
# Apply the analysis on each dataset
process_and_analyze(benin_data, "Benin - Malanville")
```
