```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Load data from CSV files (replace filenames with your actual paths)
benin_data = pd.read_csv("benin-malanville.csv")
sierra_leone_data = pd.read_csv("sierraleone-bumbuna.csv")
togo_data = pd.read_csv("togo-dapaong_qc.csv")

# Function to clean data
def clean_data(dataframe):
    # Replace negative values with NaN, then fill NaN with the column's mean
    for col in dataframe.select_dtypes(include=[np.number]).columns:
        dataframe[col] = dataframe[col].apply(lambda x: np.nan if x < 0 else x)
        dataframe[col].fillna(dataframe[col].mean(), inplace=True)
    return dataframe

# Clean the data
benin_data = clean_data(benin_data)
sierra_leone_data = clean_data(sierra_leone_data)
togo_data = clean_data(togo_data)

# Summary statistics
def summary_statistics(dataframe):
    return dataframe.describe()

# Get summary statistics
benin_summary = summary_statistics(benin_data)
sierra_leone_summary = summary_statistics(sierra_leone_data)
togo_summary = summary_statistics(togo_data)
```
